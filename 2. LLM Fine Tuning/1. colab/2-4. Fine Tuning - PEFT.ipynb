## 2. ëª¨ë¸ ë° í† í¬ë‚˜ì´ì € ë¡œë“œ

```python
# 2. ëª¨ë¸ ë° í† í¬ë‚˜ì´ì € ë¡œë“œ: "LGAI-EXAONE/EXAONE-3.5-2.4B-Instruct" ëª¨ë¸ê³¼ í•´ë‹¹ í† í¬ë‚˜ì´ì €ë¥¼ ë¡œë“œí•©ë‹ˆë‹¤.
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch

model_id = "LGAI-EXAONE/EXAONE-3.5-2.4B-Instruct"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map={"":0}, # Ensure the model is loaded on the GPU
    trust_remote_code=True
)
model.config.use_cache = False
model.config.pretraining_tp = 1 # Not needed for this model, but good practice

tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token # Or use a specific pad token if available in the tokenizer config
```

## 3. ë°ì´í„°ì…‹ ë¡œë“œ ë° ì „ì²˜ë¦¬

```python
# 3. ë°ì´í„°ì…‹ ë¡œë“œ ë° ì „ì²˜ë¦¬: NSMC ë°ì´í„°ì…‹ì„ ë¡œë“œí•˜ê³  ëª¨ë¸ í•™ìŠµì— ì í•©í•œ í˜•íƒœë¡œ ì „ì²˜ë¦¬í•©ë‹ˆë‹¤.
from datasets import load_dataset, Dataset
import pandas as pd

# NSMC ë°ì´í„°ì…‹ íŒŒì¼ URL
train_data_url = "https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt"
test_data_url = "https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt"

try:
    # ë°ì´í„° ë‹¤ìš´ë¡œë“œ ë° pandas DataFrameìœ¼ë¡œ ë¡œë“œ
    train_df = pd.read_csv(train_data_url, sep='\t').dropna()
    test_df = pd.read_csv(test_data_url, sep='\t').dropna()

    # pandas DataFrameì„ Hugging Face Datasetìœ¼ë¡œ ë³€í™˜
    train_dataset = Dataset.from_pandas(train_df)
    test_dataset = Dataset.from_pandas(test_df)

    # í•™ìŠµì— ì‚¬ìš©í•  ë°ì´í„°ì…‹ ì„ íƒ (ì—¬ê¸°ì„œëŠ” í•™ìŠµ ë°ì´í„°ì…‹ì˜ ì¼ë¶€ë¥¼ ì‚¬ìš©)
    # ë©”ëª¨ë¦¬ ì œì•½ì„ ê³ ë ¤í•˜ì—¬ ì‘ì€ í¬ê¸°ì˜ ë°ì´í„°ì…‹ ì‚¬ìš©
    dataset = train_dataset.select(range(1000)) # í•™ìŠµ ë°ì´í„°ì…‹ì˜ ì²˜ìŒ 1000ê°œ ì˜ˆì œ ì‚¬ìš©

    print("NSMC dataset loaded.")
    print(dataset)

    # Function to format the dataset for the model
    # NSMC ë°ì´í„°ì…‹ì€ 'document'ì™€ 'label' ì»¬ëŸ¼ì„ ê°€ì§‘ë‹ˆë‹¤.
    def format_dataset(example):
        # 'document'ì™€ 'label'ì„ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ ì…ë ¥ í˜•ì‹ì— ë§ê²Œ í¬ë§·íŒ…
        # ì˜ˆ: ê¸ì •/ë¶€ì • ë¦¬ë·° ë¶„ë¥˜ íƒœìŠ¤í¬ì— ë§ê²Œ í˜•ì‹ì„ ì¡°ì •í•©ë‹ˆë‹¤.
        # ì—¬ê¸°ì„œëŠ” ê°„ë‹¨í•˜ê²Œ 'document'ì™€ 'label'ì„ í•©ì³ í…ìŠ¤íŠ¸ ìƒì„± í˜•íƒœë¡œ ë§Œë“­ë‹ˆë‹¤.
        # ì‹¤ì œ fine-tuning íƒœìŠ¤í¬ì— ë”°ë¼ ì´ ë¶€ë¶„ì€ ìˆ˜ì •ì´ í•„ìš”í•©ë‹ˆë‹¤.
        label_text = "ê¸ì •" if example['label'] == 1 else "ë¶€ì •"
        return {"text": f"ë¦¬ë·°: {example['document']}\níŒë³„: {label_text}"}


    # Apply formatting and tokenize
    dataset = dataset.map(format_dataset)

    def tokenize_function(examples):
        # Ensure padding and truncation
        return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=512) # Adjust max_length as needed

    tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=dataset.column_names)
    print("Dataset loaded and preprocessed.")
    print(tokenized_dataset)

except Exception as e:
    print(f"Error loading or processing NSMC dataset: {e}")
    dataset = None
    print("Could not load NSMC dataset. Please check the URLs and your internet connection.")


if not dataset:
     print("No dataset loaded. Please ensure the dataset is loaded correctly to proceed with training.")
```

## 4. Qlora ì„¤ì •

```python
# 4. Qlora ì„¤ì •: peft ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•˜ì—¬ QLoRA ì„¤ì •ì„ ì •ì˜í•©ë‹ˆë‹¤.
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training

if dataset: # Only proceed if a dataset was successfully loaded
    # Prepare model for k-bit training
    model = prepare_model_for_kbit_training(model)

    # Define LoRA configuration
    lora_config = LoraConfig(
        r=16, # LoRA attention dimension
        lora_alpha=16, # Alpha parameter for LoRA scaling
        lora_dropout=0.05, # Dropout probability for LoRA layers
        bias="none", # Bias type: 'none', 'all', or 'lora_only'
        task_type="CAUSAL_LM", # Task type
        target_modules=["q_proj", "k_proj", "v_proj", "o_proj"], # Modules to apply LoRA to
    )

    # Get PEFT model
    model = get_peft_model(model, lora_config)

    print("QLoRA configuration applied.")
    model.print_trainable_parameters()
else:
    print("Skipping QLoRA configuration as no dataset was loaded.")
```

## 5. ëª¨ë¸ í•™ìŠµ

```python
# 5. ëª¨ë¸ í•™ìŠµ: ì„¤ì •ëœ QLoRAì™€ ë°ì´í„°ì…‹ì„ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ì„ í•™ìŠµí•©ë‹ˆë‹¤.
from transformers import TrainingArguments

if dataset: # Only proceed if a dataset was successfully loaded
    training_arguments = TrainingArguments(
        output_dir="./results", # Output directory
        num_train_epochs=1, # Number of training epochs
        per_device_train_batch_size=4, # Batch size per device during training
        gradient_accumulation_steps=1, # Number of updates steps to accumulate the gradients for
        optim="paged_adamw_32bit", # Optimizer to use
        logging_steps=10, # Log every x updates steps
        learning_rate=2e-4, # Learning rate
        weight_decay=0.001, # Weight decay
        fp16=False, # Use mixed precision
        bf16=True, # Use bfloat16 precision
        max_grad_norm=0.3, # Maximum gradient normal
        warmup_ratio=0.03, # Ratio of warmup steps for learning rate scheduler
        lr_scheduler_type="constant", # Learning rate scheduler
        report_to="none", # Reporting to experiment tracking platforms
        save_strategy="epoch", # Save the model checkpoint every epoch
    )

    from trl import SFTTrainer

    trainer = SFTTrainer(
        model=model,
        train_dataset=tokenized_dataset,
        peft_config=lora_config,
        # dataset_text_field="text", # Removed as it's not a recognized argument
        # max_seq_length=512, # Maximum sequence length - Removed as it's not a recognized argument
        # tokenizer=tokenizer, # Removed as it's not a recognized argument
        args=training_arguments,
        # packing=False, # Pack multiple examples into one sequence - Removed as it's not a recognized argument
    )

    # Start training
    print("Starting model training...")
    trainer.train()
    print("Training finished.")
else:
     print("Skipping model training as no dataset was loaded.")
```

```python
# í…ìŠ¤íŠ¸ ìƒì„±ì„ ìœ„í•œ í•¨ìˆ˜
def generate_text(model, tokenizer, prompt, max_length=100, temperature=0.7, top_p=0.9, num_return_sequences=1):
    # ì…ë ¥ í…ìŠ¤íŠ¸ í† í¬ë‚˜ì´ì§•
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    
    # í…ìŠ¤íŠ¸ ìƒì„±
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_length=max_length,
            temperature=temperature,
            top_p=top_p,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
            repetition_penalty=1.1,
            num_return_sequences=num_return_sequences
        )
    
    # ìƒì„±ëœ í…ìŠ¤íŠ¸ë“¤ ë””ì½”ë”©
    generated_texts = []
    for output in outputs:
        generated_text = tokenizer.decode(output, skip_special_tokens=True)
        # ì…ë ¥ í”„ë¡¬í”„íŠ¸ ë¶€ë¶„ ì œê±°í•˜ê³  ìƒì„±ëœ ë¶€ë¶„ë§Œ ë°˜í™˜
        generated_text = generated_text[len(prompt):].strip()
        generated_texts.append(generated_text)
    
    return generated_texts
```

```python
# ë¦¬ë·°ì˜ ê°ì •ì„ ë¶„ì„í•˜ëŠ” í•¨ìˆ˜
def evaluate_sentiment(review, model, tokenizer):
    prompt = f"ë¦¬ë·°: {review}\níŒë³„:"
    results = generate_text(model, tokenizer, prompt, max_length=20, temperature=0.3)
    
    # ê²°ê³¼ì—ì„œ ê°ì • ë¶„ì„
    result_text = results[0].lower()
    if "ê¸ì •" in result_text or "ì¢‹" in result_text or "ë§Œì¡±" in result_text:
        return "ê¸ì •"
    elif "ë¶€ì •" in result_text or "ë‚˜ì˜" in result_text or "ë³„ë¡œ" in result_text:
        return "ë¶€ì •"
    else:
        return "ì¤‘ë¦½"
```

### í•¨ìˆ˜ ì •ì˜

```python
import time
from typing import List

def batch_inference(model, tokenizer, texts: List[str], batch_size: int = 4):
    results = []
    
    print(f"ë°°ì¹˜ í¬ê¸° {batch_size}ë¡œ {len(texts)}ê°œ í…ìŠ¤íŠ¸ ì²˜ë¦¬ ì¤‘...")
    start_time = time.time()
    
    # í…ìŠ¤íŠ¸ë¥¼ ë°°ì¹˜ ë‹¨ìœ„ë¡œ ë‚˜ëˆ„ì–´ ì²˜ë¦¬
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]
        batch_prompts = [f"ë¦¬ë·°: {text}\níŒë³„:" for text in batch_texts]
        
        # ë°°ì¹˜ í† í¬ë‚˜ì´ì§•
        batch_inputs = tokenizer(
            batch_prompts, 
            return_tensors="pt", 
            padding=True, 
            truncation=True, 
            max_length=512
        ).to(model.device)
        
        # ë°°ì¹˜ ì¶”ë¡ 
        with torch.no_grad():
            batch_outputs = model.generate(
                **batch_inputs,
                max_length=batch_inputs['input_ids'].shape[1] + 20,  # ì…ë ¥ ê¸¸ì´ + 20í† í°
                temperature=0.3,
                do_sample=True,
                pad_token_id=tokenizer.eos_token_id,
                eos_token_id=tokenizer.eos_token_id,
                repetition_penalty=1.1
            )
        
        # ë°°ì¹˜ ê²°ê³¼ ì²˜ë¦¬
        for j, output in enumerate(batch_outputs):
            generated_text = tokenizer.decode(output, skip_special_tokens=True)
            # ì…ë ¥ í”„ë¡¬í”„íŠ¸ ë¶€ë¶„ ì œê±°
            prompt_length = len(batch_prompts[j])
            generated_text = generated_text[prompt_length:].strip()
            
            # ê°ì • ë¶„ì„
            sentiment = "ê¸ì •" if any(word in generated_text.lower() for word in ["ê¸ì •", "ì¢‹", "ë§Œì¡±"]) else "ë¶€ì •"
            results.append({
                'text': batch_texts[j],
                'generated': generated_text,
                'sentiment': sentiment
            })
        
        print(f"   ì²˜ë¦¬ ì™„ë£Œ: {min(i + batch_size, len(texts))}/{len(texts)}")
    
    end_time = time.time()
    processing_time = end_time - start_time
    
    print(f"ì´ ì²˜ë¦¬ ì‹œê°„: {processing_time:.2f}ì´ˆ")
    print(f"í‰ê·  ì²˜ë¦¬ ì†ë„: {len(texts)/processing_time:.2f} í…ìŠ¤íŠ¸/ì´ˆ")
    
    return results
```

```python
# ë‘ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ ë¹„êµí•˜ëŠ” í•¨ìˆ˜
def compare_models(original_model, original_tokenizer, peft_model, peft_tokenizer, test_texts):
    print("\nëª¨ë¸ ì„±ëŠ¥ ë¹„êµ ì‹œì‘...")
    print("=" * 60)
    
    results = {
        'original': {'predictions': [], 'times': []},
        'peft': {'predictions': [], 'times': []}
    }
    
    for i, text in enumerate(test_texts):
        print(f"\ní…ŒìŠ¤íŠ¸ {i+1}: {text}")
        
        # ì›ë³¸ ëª¨ë¸ í…ŒìŠ¤íŠ¸
        print("   ì›ë³¸ ëª¨ë¸ ì¶”ë¡  ì¤‘...")
        start_time = time.time()
        original_pred = evaluate_sentiment(text, original_model, original_tokenizer)
        original_time = time.time() - start_time
        
        # PEFT ëª¨ë¸ í…ŒìŠ¤íŠ¸
        print("   PEFT ëª¨ë¸ ì¶”ë¡  ì¤‘...")
        start_time = time.time()
        peft_pred = evaluate_sentiment(text, peft_model, peft_tokenizer)
        peft_time = time.time() - start_time
        
        # ê²°ê³¼ ì €ì¥
        results['original']['predictions'].append(original_pred)
        results['original']['times'].append(original_time)
        results['peft']['predictions'].append(peft_pred)
        results['peft']['times'].append(peft_time)
        
        print(f"   ì›ë³¸: {original_pred} ({original_time:.3f}ì´ˆ)")
        print(f"   PEFT: {peft_pred} ({peft_time:.3f}ì´ˆ)")
        print(f"   ì¼ì¹˜: {'âœ…' if original_pred == peft_pred else 'âŒ'}")
    
    return results
```

```python
# ê²°ê³¼ë¥¼ ë¶„ì„í•˜ê³  í†µê³„ë¥¼ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜
def analyze_results(results):
    print("\nì„±ëŠ¥ ë¶„ì„ ê²°ê³¼")
    print("=" * 60)
    
    # ì˜ˆì¸¡ ì¼ì¹˜ë„ ê³„ì‚°
    matches = sum(1 for orig, peft in zip(results['original']['predictions'], results['peft']['predictions']) if orig == peft)
    total = len(results['original']['predictions'])
    agreement = matches / total * 100
    
    # í‰ê·  ì¶”ë¡  ì‹œê°„ ê³„ì‚°
    avg_original_time = sum(results['original']['times']) / len(results['original']['times'])
    avg_peft_time = sum(results['peft']['times']) / len(results['peft']['times'])
    
    # ì˜ˆì¸¡ ë¶„í¬ ê³„ì‚°
    original_pos = sum(1 for p in results['original']['predictions'] if p == 'ê¸ì •')
    original_neg = sum(1 for p in results['original']['predictions'] if p == 'ë¶€ì •')
    peft_pos = sum(1 for p in results['peft']['predictions'] if p == 'ê¸ì •')
    peft_neg = sum(1 for p in results['peft']['predictions'] if p == 'ë¶€ì •')
    
    print(f"ì˜ˆì¸¡ ì¼ì¹˜ë„: {agreement:.1f}% ({matches}/{total})")
    print(f"í‰ê·  ì¶”ë¡  ì‹œê°„:")
    print(f"   ì›ë³¸ ëª¨ë¸: {avg_original_time:.3f}ì´ˆ")
    print(f"   PEFT ëª¨ë¸: {avg_peft_time:.3f}ì´ˆ")
    print(f"   ì†ë„ í–¥ìƒ: {avg_original_time/avg_peft_time:.2f}x")
    
    print(f"\nì˜ˆì¸¡ ë¶„í¬:")
    print(f"   ì›ë³¸ ëª¨ë¸ - ê¸ì •: {original_pos}ê°œ, ë¶€ì •: {original_neg}ê°œ")
    print(f"   PEFT ëª¨ë¸   - ê¸ì •: {peft_pos}ê°œ, ë¶€ì •: {peft_neg}ê°œ")
    
    # ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰ ë¹„êµ (ëŒ€ëµì )
    print(f"\në©”ëª¨ë¦¬ íš¨ìœ¨ì„±:")
    print(f"   PEFTëŠ” ì›ë³¸ ëª¨ë¸ ëŒ€ë¹„ í›¨ì”¬ ì ì€ ë©”ëª¨ë¦¬ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.")
    print(f"   (PEFT ì–´ëŒ‘í„°ë§Œ ì €ì¥í•˜ë¯€ë¡œ ì „ì²´ ëª¨ë¸ í¬ê¸°ì˜ 0.2% ìˆ˜ì¤€)")
```

## 11. ì¶”ë¡  ìµœì í™”
- ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ ì¶”ë¡  ë°©ë²•ë“¤

### í•¨ìˆ˜ ì •ì˜ 

```python
import gc
import psutil
import torch
from contextlib import contextmanager

# í˜„ì¬ ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰ì„ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜
def get_memory_usage():
    
    process = psutil.Process()
    memory_info = process.memory_info()
    return memory_info.rss / 1024 / 1024  # MB ë‹¨ìœ„
```

```python
# ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ ì¶”ë¡ ì„ ìœ„í•œ ì»¨í…ìŠ¤íŠ¸ ë§¤ë‹ˆì €
@contextmanager
def memory_efficient_inference(model, tokenizer):
    
    print("ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ ì¶”ë¡  ëª¨ë“œ í™œì„±í™”...")
    
    # ì¶”ë¡  ì „ ë©”ëª¨ë¦¬ ì •ë¦¬
    torch.cuda.empty_cache() if torch.cuda.is_available() else None
    gc.collect()
    
    initial_memory = get_memory_usage()
    print(f"   ì´ˆê¸° ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰: {initial_memory:.1f} MB")
    
    try:
        yield model, tokenizer
    finally:
        # ì¶”ë¡  í›„ ë©”ëª¨ë¦¬ ì •ë¦¬
        torch.cuda.empty_cache() if torch.cuda.is_available() else None
        gc.collect()
        
        final_memory = get_memory_usage()
        print(f"   ìµœì¢… ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰: {final_memory:.1f} MB")
        print(f"   ë©”ëª¨ë¦¬ ë³€í™”: {final_memory - initial_memory:+.1f} MB")
```

```python
# ìµœì í™”ëœ í…ìŠ¤íŠ¸ ìƒì„± í•¨ìˆ˜
def optimized_generate_text(model, tokenizer, prompt, max_length=50, temperature=0.3):
    
    # ì…ë ¥ í† í¬ë‚˜ì´ì§• (ìµœì†Œí•œì˜ í† í°ë§Œ ì‚¬ìš©)
    inputs = tokenizer(
        prompt, 
        return_tensors="pt", 
        truncation=True, 
        max_length=256  # ì…ë ¥ ê¸¸ì´ ì œí•œ
    ).to(model.device)
    
    # ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ ìƒì„± ì„¤ì •
    generation_config = {
        "max_length": inputs['input_ids'].shape[1] + max_length,
        "temperature": temperature,
        "top_p": 0.9,
        "do_sample": True,
        "pad_token_id": tokenizer.eos_token_id,
        "eos_token_id": tokenizer.eos_token_id,
        "repetition_penalty": 1.1,
        "no_repeat_ngram_size": 2,  # ë°˜ë³µ ë°©ì§€
        "early_stopping": True,     # ì¡°ê¸° ì¢…ë£Œ
    }
    
    with torch.no_grad():
        # ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰ ëª¨ë‹ˆí„°ë§
        start_memory = get_memory_usage()
        
        outputs = model.generate(**inputs, **generation_config)
        
        end_memory = get_memory_usage()
        print(f"   ìƒì„± ì¤‘ ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰: {end_memory - start_memory:+.1f} MB")
    
    # ê²°ê³¼ ë””ì½”ë”©
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    generated_text = generated_text[len(prompt):].strip()
    
    return generated_text
```

```python
# ìŠ¤íŠ¸ë¦¬ë° ë°©ì‹ìœ¼ë¡œ ëŒ€ëŸ‰ì˜ í…ìŠ¤íŠ¸ë¥¼ ì²˜ë¦¬í•˜ëŠ” í•¨ìˆ˜
def stream_inference(model, tokenizer, texts, chunk_size=5):
    
    print(f"ìŠ¤íŠ¸ë¦¬ë° ì¶”ë¡  ì‹œì‘ (ì²­í¬ í¬ê¸°: {chunk_size})")
    
    results = []
    total_chunks = (len(texts) + chunk_size - 1) // chunk_size
    
    for i in range(0, len(texts), chunk_size):
        chunk_texts = texts[i:i + chunk_size]
        chunk_num = i // chunk_size + 1
        
        print(f"\nì²­í¬ {chunk_num}/{total_chunks} ì²˜ë¦¬ ì¤‘...")
        
        # ê° ì²­í¬ë¥¼ ê°œë³„ì ìœ¼ë¡œ ì²˜ë¦¬í•˜ì—¬ ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰ ìµœì†Œí™”
        with memory_efficient_inference(model, tokenizer):
            for text in chunk_texts:
                prompt = f"ë¦¬ë·°: {text}\níŒë³„:"
                result = optimized_generate_text(model, tokenizer, prompt)
                
                # ê°„ë‹¨í•œ ê°ì • ë¶„ì„
                sentiment = "ê¸ì •" if any(word in result.lower() for word in ["ê¸ì •", "ì¢‹", "ë§Œì¡±"]) else "ë¶€ì •"
                
                results.append({
                    'text': text,
                    'result': result,
                    'sentiment': sentiment
                })
        
        # ì²­í¬ ê°„ ë©”ëª¨ë¦¬ ì •ë¦¬
        torch.cuda.empty_cache() if torch.cuda.is_available() else None
        gc.collect()
    
    return results
```

```python
# ë‹¤ì–‘í•œ ì¶”ë¡  ë°©ë²•ì˜ ì„±ëŠ¥ì„ ë²¤ì¹˜ë§ˆí¬í•˜ëŠ” í•¨ìˆ˜
def benchmark_inference_methods(model, tokenizer, test_texts):
    
    print("âš¡ ì¶”ë¡  ë°©ë²•ë³„ ì„±ëŠ¥ ë²¤ì¹˜ë§ˆí¬")
    print("=" * 50)
    
    methods = {
        "ì¼ë°˜ ì¶”ë¡ ": lambda: [evaluate_sentiment(text, model, tokenizer) for text in test_texts],
        "ìµœì í™”ëœ ì¶”ë¡ ": lambda: [optimized_generate_text(model, tokenizer, f"ë¦¬ë·°: {text}\níŒë³„:") for text in test_texts],
        "ìŠ¤íŠ¸ë¦¬ë° ì¶”ë¡ ": lambda: stream_inference(model, tokenizer, test_texts)
    }
    
    results = {}
    
    for method_name, method_func in methods.items():
        print(f"\n{method_name} í…ŒìŠ¤íŠ¸ ì¤‘...")
        
        start_time = time.time()
        start_memory = get_memory_usage()
        
        try:
            if method_name == "ìŠ¤íŠ¸ë¦¬ë° ì¶”ë¡ ":
                method_results = method_func()
                predictions = [r['sentiment'] for r in method_results]
            else:
                predictions = method_func()
            
            end_time = time.time()
            end_memory = get_memory_usage()
            
            results[method_name] = {
                'time': end_time - start_time,
                'memory_delta': end_memory - start_memory,
                'predictions': predictions
            }
            
            print(f"   ì‹œê°„: {end_time - start_time:.2f}ì´ˆ")
            print(f"   ë©”ëª¨ë¦¬ ë³€í™”: {end_memory - start_memory:+.1f} MB")
            print(f"   ì²˜ë¦¬ëŸ‰: {len(test_texts)/(end_time - start_time):.1f} í…ìŠ¤íŠ¸/ì´ˆ")
            
        except Exception as e:
            print(f"   âŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
            results[method_name] = {'error': str(e)}
    
    return results
```
