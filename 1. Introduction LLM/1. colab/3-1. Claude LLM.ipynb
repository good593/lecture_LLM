# [클로드 3(Claude 3)](https://aiheroes.ai/community/150)
- 클로드 3는 2024년 3월 4일에 출시된 최신의 가장 진보된 대규모 언어 모델로, Anthropic에서 개발했습니다.
- Claude 시리즈를 만든 Anthropic사는 OpenAI의 개발자들이 만든 회사입니다.
-  Anthropic은 '헌법적 AI'라는 독특한 접근 방식을 택했습니다.
  - Anthropic은 인간에게 도움이 되면서도 해를 끼치지 않는 안전한 인공지능을 만드는 것을 목표로 삼았습니다.
  - 이를 위해 클로드는 인간을 위협하지 않는 10가지 원칙을 바탕으로 설계되었습니다.
  - 클로드는 대화할 때마다 이 원칙들을 잘 지키고 있는지 스스로 점검하는 과정을 거치게 됩니다.

## 클로드 모델 3가지
- 클로드에는 3가지 모델이 있습니다. 1. `Haiku`, 2. `Sonnet`, 3. `Opus` 모델이 있으며, 위 이미지처럼 성능, 비용 등이 우상단으로 갈 수록 높아집니다.

image.png

### `Claude Haiku`
- 가볍고 빠른 속도를 가지며, 비용 효율성을 고려한 모델입니다.
- 주로 작업의 빠른 완료와 비용 효율성이 중요한 경우에 적합합니다.

### `Claude Sonnet`
- 우리가 로그인 시 `무료`로 사용할 수 있는 버전은 Sonnet 모델이며 GPT-3.5 보다 벤치마크 성능이 조금 우수하다고 합니다.
- 클로드 모델 시리즈 중 중간 수준의 성능을 제공하는 모델입니다.
- 기업 환경에서 데이터 처리, 품질 관리, 제품 추천 등에 활용될 수 있습니다.

### `Claude Opus`
- `유료`로 사용할 수 있는 Opus 프로 버전은 GPT-4.0 보다 조금 더 우수하게 나왔다고 합니다.
- 클로드 3중에서 가장 뛰어난 인지 능력을 갖춘 모델입니다.
- 수학적 추론부터 복잡한 천문학 계산까지, 전문적인 작업에 특화된 성능입니다.

## 다른 LLM 성능비교
- Anthropic은 Claude 3, ChatGPT, Gemini 등 생성형 AI 모델의 성능을 자체적으로 테스트했습니다.
- 학부 수준의 전문 지식(MMLU), 대학원 수준의 전문가 추론(GPQA), 기본 수학(GSM8K), 코드 등 총 10개 항목에 대해 평가한 결과, 클로드 3 Opus 모델이 모든 항목에서 가장 높은 점수를 받았습니다.
- 특히 분석, 예측, 코드 생성, 영어가 아닌 언어로 대화하기 태스크에서 Claude 3가 두각을 나타냈습니다.

## [클로드 3의 장점](https://textcortex.com/ko/post/advantages-of-claude-3)

### 텍스트 생성
- Claude 3 모델의 영어 및 비영어권 언어에 대한 문법, 철자, 문장 구조 지식 덕분에 Anthropic에서 지원하는 언어로 모든 유형의 텍스트를 생성할 수 있습니다.
- Claude 3 하이쿠 모델은 간단하고 기본적인 텍스트 생성에 적합하지만, 복잡한 글쓰기 작업에는 Claude 3 소네트 또는 오푸스 모델을 사용하는 것이 좋습니다.

image.png

### 코딩
- Claude 3 모델은 Python, C++, C#, JavaScript, SQL 등 널리 사용되는 코딩 언어를 사용하여 학습되었으므로 이러한 언어 중 어떤 언어로든 출력을 생성할 수 있습니다.
- Claude 3 모델에는 코드 실행 기능이 없지만, 사용자가 선택한 언어로 초안을 생성하여 명령의 기초로 사용할 수 있습니다.

image.png

- 또한 Claude 3 모델은 코드를 생성할 뿐만 아니라 코드 개선을 위한 권장 사항도 제공합니다.
- 예를 들어, Claude 3에게 Elden Ring과 같은 통계 시스템을 Python 언어로 작성하도록 요청하면 기본 코드 외에 추가할 수 있는 개선 사항을 나열해 줍니다.

image.png

### 컨텍스트 창
- Claude 2 모델부터 Anthropic은 사용자에게 큰 크기의 컨텍스트 창을 제공하여 대량의 데이터 작업이 필요한 작업에 이상적인 솔루션으로 자리 잡았습니다.
- Claude 3 모델은 이전 모델보다 컨텍스트 창 크기를 두 배로 늘려 사용자에게 약 15만 단어에 해당하는 200만 토큰의 컨텍스트 창을 제공합니다.
- 또한, 클로드 3 오퍼스 모델은 특정 사용 사례를 위해 최대 100만 토큰의 입력 크기를 지원합니다.
- Claude 3 모델은 주어진 데이터를 분석하고 지식을 불러오는 데 효율적이고 효과적입니다.

## 예제

    model="claude-3-5-sonnet-20241022",

                }
            ]
        }
    ]
)

## [text-generation](https://huggingface.co/mlfoundations-dev/oh-dcft-v3.1-claude-3-5-haiku-20241022)

````python
from transformers import pipeline
import torch

model_id = "mlfoundations-dev/oh-dcft-v3.1-claude-3-5-haiku-20241022"

pipe = pipeline(
    task="text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device="cuda",
)
````

````python
messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in Korean!"},
    {"role": "user", "content": "너는 누구니?"},
]

terminators = [
    pipe.tokenizer.eos_token_id,
    pipe.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]
````

````python
outputs = pipe(
    messages,
    max_new_tokens=256,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)
````

````python
assistant_response = outputs[0]["generated_text"][-1]["content"]
print(assistant_response)
````
