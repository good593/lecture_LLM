# SciSpace
- SciSpace는 논문 작성과 연구 과정을 자동화하고 효율성을 높일 수 있는 AI 도구이다.
- 주로 논문 요약과ㅘ 번역 기능을 제공하며, 연구자들이 빠르게 논문을 이해하고 정리할 수 있도록 도와준다.
- 뿐만 아니라, 논문 작성 보조 기능을 통해 논문의 구조를 잡고 더 나은 문장을 작성할 수 있도록 해준다.

## [논문작성에 유용한 생성형 AI 도구들](https://ctl.skku.edu/ctl/data.do?mode=view&articleNo=50392&article.offset=0&articleLimit=10#/list)

### [SciSpace](https://typeset.io)
 - SciSpace는 사용자가 관심있는 연구 논문(2억 8,200만 건)을 찾고, 논문에 대한 질문을 하며 다양한 논문에서 통찰력을 빠르게 얻는 데 도움이 되는 AI 기반 플랫폼입니다.
 - 특히, Scispace는 다른 생성형 AI 도구 보다 ‘논문’에 특화되어 있어 논문 요약과 해석에 대해 더 전문적이고 정확한 정보를 제공합니다.
 - 즉, 대학원생분들이 관심있는 논문을 검색하고 이를 빠르게 파악하기 위한 요약 정보를 원할 때 유용하게 사용할 수 있는 생성형 AI 도구입니다.

image.png

### [Connected Papers](https://www.connectedpapers.com/)
- Connected Papers는 연구자가 자신의 연구 분야와 관련된 논문을 찾고 탐색하는 데 도움을 주는 시각화 도구입니다.
- 연구자는 논문 간의 연결을 시각적으로 한 눈에 이해할 수 있기 때문에 본격적인 연구 시작 전 큰 그림을 그리고 연구 전반을 조망하는데 유리합니다.
- 활용 방법은 다음과 같습니다.
  - Connected Papers 검색창에 일반적인 논문을 검색하면 AI가 해당 논문과의 상호 연결을 보여주는 그래프를 생성합니다.
  - 이를 통해 새로운 논문에 쉽게 접근하고 관심 있는 분야의 트렌드, 인기 작품 및 역학에 대한 실제적이고 시각적인 이해를 얻을 수 있습니다.

### [Consensus](https://consensus.app/)
- 컨센서스(Consensus)는 논문 자료 검색과 인용에 최적화된 생성형 AI입니다.
- 컨센서스가 제공하는 주된 강점 중 하나는 방대한 데이터베이스입니다.
- 현재 컨센서스는 시맨틱 스콜라(Semantic Scholar) 데이터베이스를 기반으로 하여 2억 개 이상의 논문을 검색할 수 있습니다.
- 이를 통해 사용자는 다양한 과학 분야에 걸쳐 실제 연구 논문과 연관된 통찰력을 얻을 수 있습니다.

## SciSpace 주요 기능

- 논문 요약
  - 방대한 논문을 빠르게 분석하고 요약하여 중요한 정보 추출 가능.
  - 논문 이외 PDF 파일도 가능.
  - 다양한 언오로 요약 가능.
- 논문 작성 보조
  - 논문의 전체 흐름과 구조를 잡는 데 도움을 줌.
- 크롬확장
  - SciSpace를 크롬에 추가해서 실시간 웹페이지 내용을 요약하거나 관련 질문 가능

# [with Huggingface](https://huggingface.co/SciSpace)

````python
from transformers import pipeline
import torch

model_id = "SciSpace/phi-3-model-3375"

pipe = pipeline(
    task="text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device="cuda",
)
````

````python
messages = [
    {"role": "system", "content": "You are a assistant who always responds in Korean!"},
    {"role": "user", "content": "RNN 모델에 대해서 간단하게 설명해줘"},
]

terminators = [
    pipe.tokenizer.eos_token_id,
    pipe.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]
````

````python
outputs = pipe(
    messages,
    max_new_tokens=1000,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)
````

````python
assistant_response = outputs[0]["generated_text"][-1]["content"]
print(assistant_response)
````
